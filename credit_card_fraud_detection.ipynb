```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from keras.layers import concatenate
```

```python
credit_card_data=pd.read_csv("creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
legit=credit_card_data[credit_card_data.Class == 0]
frawd=credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(frawd.shape)
```

```python
legit.Amount.describe()
```

```python
frawd.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=93)
```

```python
new_dataset=pd.concat([legit_sample,frawd],axis=0)
```

```python
print(new_dataset)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction= model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data :", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score on TEst Data : ", training_data_accuracy )
```
